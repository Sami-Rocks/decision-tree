```python
def get_numbers():
    number = numbers[random.randrange(len(numbers))]
    recipient_list = [numbers[random.randrange(len(numbers))],
                      numbers[random.randrange(len(numbers))], 
                      numbers[random.randrange(len(numbers))],
                      numbers[random.randrange(len(numbers))],
                      numbers[random.randrange(len(numbers))]]
    fraud_list = [numbers[random.randrange(len(numbers))],
                      numbers[random.randrange(len(numbers))], 
                      numbers[random.randrange(len(numbers))]]
    sender = numbers[random.randrange(len(numbers))]

    return recipient_list, fraud_list, sender
```
