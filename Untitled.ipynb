```python
def randTime():
    ranH = random.randint(0,10)
    ranM = random.randint(10,59)
    ranS = random.randint(10,59)
    ranM = random.randint(1, 12)
    ranD = random.randint(1, 30)
    
    if(ranM <= 2):
        SranM = ranM +50
    else:
        SranM = ranM - 2 
        
        
    timeS = datetime(2019, ranM, ranD, ranH, SranM, ranS, 0)
    timeS_year = timeS.year
    timeS_month = timeS.month
    timeS_day = timeS.day
    timeS_hour = timeS.hour
    timeS_minute =timeS.minute
    timeS_second = timeS.second
    
    timeE = datetime(2019, ranM, ranD, ranH, ranM, ranS, 0)
    timeE_year = timeE.year
    timeE_month = timeE.month
    timeE_day = timeE.day
    time_hour = timeE.hour
    timeE_minute = timeE.minute
    timeE_second = timeE.second
    
    
    return ( timeS_year,
             timeS_month,
             timeS_day,
             timeS_hour,
             timeS_minute,
             timeS_second,
             timeE_year,
             timeE_month,
             timeE_day,
             time_hour,
             timeE_minute,
             timeE_second)
```

```python
def get_numbers():
    number = str(numbers[random.randrange(len(numbers))]).strip('[]')
    recipient_list = [(str(numbers[random.randrange(len(numbers))]).strip('[]')),
                      (str(numbers[random.randrange(len(numbers))]).strip('[]')), 
                      (str(numbers[random.randrange(len(numbers))]).strip('[]')),
                      (str(numbers[random.randrange(len(numbers))]).strip('[]')),
                      (str(numbers[random.randrange(len(numbers))]).strip('[]'))]
    fraud_list = [(str(numbers[random.randrange(len(numbers))]).strip('[]')),
                      (str(numbers[random.randrange(len(numbers))]).strip('[]')), 
                      (str(numbers[random.randrange(len(numbers))]).strip('[]'))]
    sender = (str(numbers[random.randrange(len(numbers))]).strip('[]'))

    return recipient_list, fraud_list, sender
```

```python
def generate_list(amt1, amt2, amt3, amt4, userID):
    recipient_list, fraud_list, sender = get_numbers()
    recipient = recipient_list[random.randrange(len(recipient_list))]
    for x in range(11):
        ran1 = random.randint(22,550)
        amount = round(random.uniform(amt1,amt2), 2)
        fraud = 'NOT_FRAUD'
        if(x == 4):
            amount = round(random.uniform(amt3,amt4), 2)
            recipient = fraud_list[random.randrange(len(fraud_list))]  
            fraud = 'FRAUD'
            
        if(x == random.randint(0,10)):
            amount = round(random.uniform(amt3,amt4), 2)
            recipient = fraud_list[random.randrange(len(fraud_list))]
            fraud = 'FRAUD'
         
        if(x == random.randint(0,10)):
            amount = round(random.uniform(amt3,amt4), 2)
            recipient = fraud_list[random.randrange(len(fraud_list))]
            fraud = 'FRAUD'
        
        timeS_year, timeS_month, timeS_day, timeS_hour, timeS_minute, timeS_second, timeE_year, timeE_month, timeE_day, time_hour, timeE_minute, timeE_second, = randTime()
        data.append([userID, 
                     ran1, 
                     amount, 
                     recipient, 
                     sender, 
                     timeS_year,
                     timeS_month,
                     timeS_day,
                     timeS_hour,
                     timeS_minute,
                     timeS_second,
                     timeE_year,
                     timeE_month,
                     timeE_day,
                     time_hour,
                     timeE_minute,
                     timeE_second,
                     fraud])
```
